```python
# 1. Instalação das ferramentas (Execute apenas uma vez no Colab)
# !pip install diffusers transformers accelerate


import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt


# 2. Carregando o Modelo (O Cérebro do Escultor)
modelo_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(modelo_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda") # Direciona para a GPU para ser rápido


# 3. Configurando o Desafio (Onde você mexe nos parâmetros)
meu_prompt = "A cinematic photo of a cat astronaut on Mars, high resolution, 299x299"
minha_guidance_scale = 15.0 # Teste mudar para 2.0 e veja a diferença!


# 4. Executando o Escultor (Denoiser + UNet)
print("Aguarde... O Denoiser está limpando o ruído...")
imagem = pipe(prompt=meu_prompt, guidance_scale=minha_guidance_scale).images[0]


# 5. Visualizando o Resultado (O Crítico de Arte)
plt.imshow(imagem)
plt.axis('off') # Remove as réguas do gráfico
plt.show()


# 6. Salvando a Obra
imagem.save("gato_astronauta.png")
print("Imagem salva como gato_astronauta.png!")
```
